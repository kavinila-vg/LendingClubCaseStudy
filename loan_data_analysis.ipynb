```python
def ratiobargraph(feature, data=df3, figsize=(10,5), rsorted=True):
    plt.figure(figsize=figsize)
    if rsorted:
        unique_values = sorted(data[feature].unique())
    else:
        unique_values = data[feature].unique()
    ratio_list = []
    for value in unique_values:
        val = data[data[feature]==value]
        count = len(val[val["loan_status"]=="Charged Off"])
        ratio_list.append(count*100/val["loan_status"].count())
   
    plt.bar(unique_values, ratio_list, color='green', edgecolor='white')
    plt.title("Loan Defaults percentage for "+str(feature)+" feature")
    plt.xlabel(feature, fontsize=16)
    plt.ylabel("defaulter percentage", fontsize=16)
    plt.show()
```

```python
def bargraph(x, figsize=(10,5)):
    plt.figure(figsize=figsize)
    sns.barplot(x=x, y='loan_status', data=df3)
    plt.title("Loan Defaults for "+str(x))
    plt.xlabel(x, fontsize=16)
    plt.ylabel("defaulter ratio", fontsize=16)
    plt.show()
```

# Bivariate analysis

### from the univariate analysis we found that loan_amnt, installment, term, purpose ,int_rate, dti, grade, sub_grade so will check the if there is any relation between these features

```python
def biivariatebargraph(x, hue):
    plt.figure(figsize=(12,8))
    sns.barplot(x=x, y='loan_status_num', hue=hue, data=df3)
    plt.title(x+" Vs "+hue+" for Loan Default ratio")
    plt.xlabel(x, fontsize=16)
    plt.ylabel("defaulter ratio", fontsize=16)
    plt.show()
```

```python
def to_numeric(x):
    if x=="Charged Off": 
        return 1 
    else: 
        return 0
```

```python
df3['loan_status_num']=df3['loan_status'].apply(to_numeric)
```
